# **S√©paration des listes de produits en ligne**

```python
# 1. S√©paration des produits : chaque produit sera sur une ligne distincte
from itertools import chain

df['Product'] = df['Product'].str.strip('[]').str.replace("'", "").str.split(',')
df = df.explode('Product')
```

```python
df
```

```python
# Ajouter la colonne Id
df = df.assign(Id=range(1, len(df) + 1))
```

```python
df
```

# **Traitement des dates**

```python
# Transformer les dates
df['Date'] = pd.to_datetime(df['Date'])

# Remplacer les valeurs NaN dans la colonne Promotion par "No Promotion"
df['Promotion'].fillna('No Promotion', inplace=True)

# Afficher les types de donn√©es apr√®s transformationa
print(df.dtypes)
```

```python
# 2. Transformation de la date
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Season'] = df['Season'].astype('category')  # Saison cat√©goris√©e

```

```python
# Affichage des colonnes sp√©cifiques
result = df[['Date', 'Year', 'Month', 'Day', 'Season']]
print(result)
```
